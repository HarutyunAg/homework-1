```python
import pandasql as ps
import pandas as pd
import sqlite3

bank_data = pd.read_excel('bank_account_data.xlsx')
```

```python
# Опишите имеющиеся колонки в данных - сделайте аналог describe в pandas для колонок.

def describe_sql(data: pd.DataFrame, feature: str) -> pd.DataFrame:
    """
    The function takes a dataframe 'data' and the name of the parameter 'feature',
    and returns the sum, the average, minimum and maximum value of the column in the form of a pd.DataFrame.
    
    :param data: pd.DataFrame
    :param feature: str
    :return: pd.DataFrame
    """
    query_task_one = f"""
    SELECT SUM({feature}),COUNT({feature}), AVG({feature}), MIN({feature}), MAX({feature})
    FROM data;
    """
    return ps.sqldf(query_task_one)


for i in bank_data.columns:
    display(describe_sql(bank_data, i))
```

```python
# Посчитайте совокупный объем средств у клиентов в банке - сумма по всем балансам.
query_task_two = """
SELECT Cust_ID, (Cheq_Bal + Sav_Bal + Debt_Bal) AS total_bal
FROM bank_data
GROUP BY Cust_ID
"""
ps.sqldf(query_task_two)
```

```python
# Выведите только тех клиентов, которые приобрели ипотеку.
query_task_tree = """
SELECT Cust_ID
FROM bank_data
WHERE Purchase = 1;
"""
ps.sqldf(query_task_tree)
```

```python
# Совокупный объем средств в среднем у клиентов, которые приобрели ипотеку
# и у клиентов, которые её не приобретали
query_task_four = """
SELECT AVG(Cheq_Bal + Sav_Bal + Debt_Bal) AS mean_total_bal
FROM bank_data
GROUP BY Purchase;
"""
ps.sqldf(query_task_four)

# Видно что совокупный объем средств в среднем у клиентов с ипотекой больше, чем у клиентов без ипотеки.
# Значение 0 - без ипотеки
# Значение 1 - с ипотекой
```

```python
# Есть ли разница в среднем по каждому из типов баланса у клиентов,
# которые приобрели ипотеку и у клиентов, которые её не приобретали?
query_task_five = """
SELECT AVG(Cheq_Bal), AVG(Sav_Bal), AVG(Debt_Bal)
FROM bank_data
GROUP BY Purchase;
"""

ps.sqldf(query_task_five)
```

```python
# Выберите самого возрастного получателя ипотеки.
query_task_six = """
SELECT Cust_ID, MAX(Age)
FROM bank_data
WHERE Purchase = 1;
"""
ps.sqldf(query_task_six)

# Самому возрастному клиенту которому одобрили ипотеку 45 лет. Его ID 41
```

```python
# Создайте любую базу данных в sqlite3, состоящую из 5 колонок,
# одна из колонок обязательно должна быть типа REAL. (Названия колонок любые, какие захотите)

connect_foo = sqlite3.connect('foo_data.db')

connect_foo.execute('''
CREATE TABLE game(
    id INTEGER, 
    player_name STRING,
    win INTEGER,
    losses INTEGER,
    winrate REAL);
    ''')

connect_foo.commit()
connect_foo.close()
```

```python
# Добавьте 3 полностью заполненные строки в вашу БД.
query_task = """
INSERT INTO game VALUES
('1', 'baka', '150', '145', '50,84'),
('2', 'mane', '95', '120', '44,18'),
('3', 'trix', '320', '280', '53,33');
"""
connect_foo = sqlite3.connect('foo_data.db')
connect_foo.execute(query_task)
connect_foo.commit()
connect_foo.close()
```

```python
connect_foo = sqlite3.connect('foo_data.db')
qwe = connect_foo.cursor()
qwe.execute("""SELECT * FROM game;""")
qwe.fetchall()
```

```python
# Удалите строку из колонки типа REAL (из пункта 8),
# в которой значение признака среднее (то есть не минимальное и не максимальное).
# Напрямую указывать значение из этой строки нельзя.
query_task_del = """
DELETE FROM game
WHERE winrate NOT IN (
    (SELECT MIN(winrate) FROM game),
    (SELECT MAX(winrate) FROM game)
);
"""
connect_foo = sqlite3.connect('foo_data.db')
connect_foo.execute(query_task_del)
connect_foo.commit()
connect_foo.close()

# Должна удалиться строка с индексом 1:
# (1, 'baka', 150, 145, '50,84')
```
