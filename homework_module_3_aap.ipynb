```python
# task reversed integer
def reverse_number(n: int) -> int:
    """
    Accepts a number, and return an integer,
    the digits in which are written in reverse order.

    :param n: int - input number
    :return: int - reversed number
    """
    str_n = str(abs(n))
    revers = str_n[::-1]

    return int(revers)


print(reverse_number(int(input())))
```

```python
# task square
def make_square(size: int):
    """
    The function will draw a square of asterisks.

    :param size: int - the size of the square in asterisks.
    """
    inner_size = size - 2
    print('*' * size)
    for i in range(inner_size):
        print('*' + ' ' * inner_size + '*')
    print('*' * size)


make_square(int(input()))
```

```python
# task permutationn
from itertools import permutations


def list_permutation(n: list) -> list:
    """
    Accepts an array as input and returns all possible permutations with its numbers.

    :param n: list
    :return: list of tuple - list contains all possible permutations.
    """
    result = []
    for i in permutations(n):
        result.append(i)
    return result


print(list_permutation([0, 1]))  # As an example.
```

```python
# task_almost_double_factorial
def almost_double_factorial(n: int) -> int:
    """
    The function outputs the product of all even natural numbers not exceeding n.
    :type n: int
    :return: int
    """
    prod = 1
    for i in range(2, n+1, 2):
        prod *= i
    return prod


N = 25  # Testing the function above
N_double_fact = almost_double_factorial(N)
assert N_double_fact == 1961990553600, "Something is wrong! Please try again"
```

```python
# task_more_strings
s_, a_, b_ = (input() for _ in range(3))


def foo_func(s: str, a: str, b: str, counter: int = 0):
    """
    The function outputs the minimum number of operations,
    after which there will be no occurrences of the string "a" in the string "s",
    or "Impossible" if more than 1000 operations are required.

    :param s: str
    :param a: str
    :param b: str
    :param counter: int - number of operations
    """
    if a in b and a in s or counter >= 1000:
        return 'Impossible'
    if s == s.replace(a, b):
        return counter
    else:
        counter += 1
        s = s.replace(a, b)
        return foo_func(s, a, b, counter)


print(foo_func(s_, a_, b_))
```

```python
#task_creating_arrays
def num_sum_and_erase(a: list, erase: int = 1) -> list:
    """
    The function create a new list (b_list) of partial sums of the list "a",
    and remove from list "B" all elements equal to the "erase" parameter.

    :param erase: int
    :param a: list
    :return: list
    """
    b_list = []
    for i in range(1, len(a) + 1):
        b = sum(a[:i])
        if b != erase:
            b_list.append(b)
    return b_list


A = [5, 1, 4, 5, 14]
B = num_sum_and_erase(A, erase=10)

print(B)
assert B == [5, 6, 15, 29], "Something is wrong! Please try again"
```

```python
#task_string_processing
sentences = ['1 thousand devils', 'My name is 9Pasha', 'Room #125 costs $100']


def str_process(str_list: list) -> list:
    """
    The function removes words from the list containing numbers and special characters.
    param str_list: list
    :return: list
    """
    result = []
    for string in str_list:
        s = string.split()
        result.append(' '.join([x for x in s if x.isalpha()]))
    return result


assert str_process(sentences) == ['thousand devils', 'My name is', 'Room costs'],\
    "Something is wrong! Please try again"
```

```python
#task_hamming_distance
def hamming(str_a: str, str_b: str) -> int:
    """
    the function calculates the Hamming distance,
    which is defined as the number of characters that do not match
    between two strings in the same position.
    :param str_b: str
    :param str_a: str
    :return: int
    """
    count = 0
    for i in range(len(str_a)):
        if str_a[i] != str_b[i]:
            count += 1
    return count


assert hamming("AGCT", "ACGT") == 2
```

```python
import numpy as np

# Для генерации матриц используем фукнцию random -- она используется для генерации случайных объектов
# функция sample создает случайную выборку. В качестве аргумента ей передается кортеж (i,j),
# здесь i -- число строк, j -- число столбцов.
a = np.random.sample((100, 100))
b = np.random.sample((100, 100))

# Выведите размерность (ранг) каждой матрицы с помощью функции ndim.
# Используйте функцию shape, что она вывела?
print(np.ndim(a))
print(np.ndim(b))

print(np.shape(a))
print(np.shape(b))


def mult(a_lst: list, b_lst: list):
    """
    a_lst: list of "size" lists, each contains "size" floats --- первая матрица-аргумент
    b_lst: list of "size" lists, each contains "size" floats --- вторая матрица-аргумент
    return c: list of "size" lists, each contains "size" floats --- матрица,
    являющаяся результатом умножения матриц a и b

    Функция принимает на вход две матрицы: a и b размерностью size x size
    Возвращает матрицу их произведения a * b = c
    Реализуйте умножение матриц без использования функций из пакета numpy (может помочь функция zip)
    """

    zip_b = zip(*b_lst)
    zip_b = list(zip_b)
    return [[sum(elem_a * elem_b for elem_a, elem_b in zip(row_a, col_b))
             for col_b in zip_b] for row_a in a_lst]


def np_mult(a_list, b_list) -> list:
    """
    a: np.array[size, size] --- первая матрица-аргумент
    b: np.array[size, size] --- вторая матрица-аргумент
    return c: np.array[size, size]       --- матрица, являющаяся результатом умножения матриц a и b

    Функция принимает на вход две матрицы: a и b размерностью size x size
    Возвращает матрицу их произведения a * b = c

    Реализуйте умножение матриц, используя функции из пакета numpy
    """

    c_list = np.dot(a_list, b_list)
    return list(c_list)


M1 = mult(a, b)
M2 = np_mult(a, b)
assert np.allclose(np.array(M1), M2)
```
