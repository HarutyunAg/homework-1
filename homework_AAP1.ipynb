```python
# task 4
numbers = input()  # Get a string of numbers.


def sum_neighbors(list_n):
    """The function returns a list. Each element is the sum of its neighbors."""
    result_list = []
    if len(list_n) == 1:
        result_list.append(list_n[0])
        return result_list

    for i in range(0, len(list_n)):
        if i == 0:
            result_list.append(list_n[i+1] + list_n[len(list_n)-1])
        elif i == len(list_n)-1:
            result_list.append(list_n[0] + list_n[i-1])
        else:
            result_list.append(list_n[i+1] + list_n[i-1])
    return result_list


list_num = [int(i) for i in numbers.split()]  # List representation.
result = sum_neighbors(list_num)

print(*result)
```

```python
# task 5
num = [int(i) for i in input().split()]  # Accepts a list of numbers in one line as input


def find_duplicate(list_num: list):
    """
    Displays on the screen in one line the values that occur in it more than once.
    """
    num_dict = dict()
    for i in list_num:
        num_dict[i] = num_dict.get(i, 0) + 1

    duplicate_num = []
    for k, v in num_dict.items():
        if v > 1:
            duplicate_num.append(k)

    print(*duplicate_num)


find_duplicate(num)
```

```python
# task 7
s = input()
index = 0
count = 1
res = ''

while index < (len(s) - 1):
    if s[index] == s[index+1]:
        count += 1
    else:
        res += s[index] + str(count)
        count = 1
    index += 1

result = res + s[index] + str(count)
print(result)  # Outputs the encoded sequence to standard output.
```

```python
# task 8
sum_n = 0
numbers = []

while True:
    """
    Reads numbers from the console (one per line)
    until the sum of the entered numbers is equal to 0.
    """
    n = int(input())
    sum_n += n
    numbers.append(n)
    if sum_n == 0:
        break

result = 0
for i in numbers:
    result += i**2

print(result)
# Outputs the sum of the squares of all the numbers read.
```

```python
# task 9
def fib(n: int) -> int:
    """
    This problem is solved using a sequence of Fibonacci numbers.


    :type n: int - is the total number of stairs (the index of the number in the Fibonacci sequence).
    :return: int - the number of unique ways to get to the top.
    """
    a = 0
    b = 1
    for i in range(n+1):
        a, b = b, a + b
    return a


stair = int(input())
print(fib(stair))
```

```python
# task 10
def find_prefix(arr: list) -> str:
    """
    Finds the longest prefix in a string list
    :type arr: list - list of strings
    :return: str - the longest prefix
    """
    size = len(arr)
    if size == 0:
        return ""

    if size == 1:
        return arr[0]

    arr.sort()
    end = min(len(arr[0]), len(arr[size - 1]))

    i = 0
    while i < end and arr[0][i] == arr[size - 1][i]:
        i += 1

    prefix = arr[0][0: i]
    return prefix


inp = input().split()
print(find_prefix(inp))
```
